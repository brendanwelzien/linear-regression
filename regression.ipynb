```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

# Show the first five rows of the data set

```python
fishy = pd.read_csv('fish.csv')
print(fishy.head(5))
```

# show the description and info the data set

```python
fishy.describe()
```

```python
fishy.info()
```

```python
X = fishy['Width'].values
y = fishy['Height'].values
X
```

```python
X_shape = X.reshape(-1, 1)
X_shape
```

# Split data intro train and test by using regression model

```python
print(X)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_shape,y,train_size=0.8,test_size=0.2,random_state=100)
print(f"X_train shape {x_train.shape}")
print(f"y_train shape {y_train.shape}")
print(f"X_test shape {x_test.shape}")
print(f"y_test shape {y_test.shape}")
print(y_test)
```

```python
%matplotlib inline
plt.scatter(x_train,y_train,color='orange')
plt.xlabel('Weight in g')
plt.ylabel('Height in cm')
plt.title('Fish Wt. Ht.')
plt.show()
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
estimate_y = lr.predict(x_test)
print(f"Train score {round(lr.score(x_train,y_train)*100, 2)} %")
print(f"Test score {round(lr.score(x_test, y_test)*100, 2)} %")
```

```python
plt.scatter(x_train,y_train,color='orange')
plt.plot(x_test,estimate_y)
plt.xlabel('Weight in g')
plt.ylabel('Height in cm')
plt.title('Fish Wt. Ht.')
plt.plot
```
